```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

# Loading Housing Data



```python
data = pd.read_csv("train.csv")
```

```python
data.head(10)
```

```python
data.info()
```

```python
data = data.select_dtypes(exclude=['object'])
```

```python
data.info()
```

```python
# Impute missing values with mean
data.fillna(data.mean(), inplace=True)
data.drop(columns=["Id"], inplace=True)
data
```

```python
data.info()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

# Assuming data_dropped_vif is your feature matrix and target variable is 'SalePrice'
X_train = data.drop(columns=["SalePrice"])
y_train = data["SalePrice"]

# Initialize Random Forest Regressor
rf_regressor = RandomForestRegressor(random_state=42)

# Fit the model
rf_regressor.fit(X_train, y_train)
```

```python
# Load the test data
test_data = pd.read_csv('test.csv')
test_data.fillna(data.mean(), inplace=True)
test_data.drop(columns=["Id"], inplace=True)


test_data = test_data.select_dtypes(exclude=['object'])
test_data
```

```python
test_data.info()
```

```python
# Predict on the filtered test set
y_pred_rf = rf_regressor.predict(test_data.values)

# Get the IDs from the test data
test = pd.read_csv('test.csv')
test_ids = test['Id']

# Create a DataFrame with IDs and predicted prices
predictions_df = pd.DataFrame({'ID': test_ids, 'SalePrice': y_pred_rf})

# Save the predictions to a CSV file
predictions_df.to_csv('Submission.csv', index=False)
```
